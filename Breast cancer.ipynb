# 5. Handling Outlier

```python
plt.figure(figsize =(8,4))
sns.boxplot(x = 'area_mean', data=df1, orient='h');
```

```python
plt.figure(figsize =(8,4))
sns.boxplot(x = 'area_worst', data=df1, orient='h');
```

```python
for colname,colvals in df1.iteritems():
    if df1[colname].dtype!=object and colname!='diagnosis':
        Q1=df1[colname].quantile(q=0.25)
        Q3=df1[colname].quantile(q=0.75)
        IQR=Q3-Q1
        low=Q1-(1.5*IQR)
        up=Q3+(1.5*IQR)
        df1.loc[(df1[colname]<=low) | (df1[colname]>=up) ,colname]=df1[colname].median()
    

plt.figure(figsize =(8,4))
sns.boxplot(x = 'area_mean', data=df1, orient='h');
```

```python
for colname,colvals in df1.iteritems():
    if df1[colname].dtype!=object and colname!='diagnosis':
        Q1=df1[colname].quantile(q=0.25)
        Q3=df1[colname].quantile(q=0.75)
        IQR=Q3-Q1
        low=Q1-(1.5*IQR)
        up=Q3+(1.5*IQR)
        df1.loc[(df1[colname]<=low) | (df1[colname]>=up) ,colname]=df1[colname].median()

plt.figure(figsize =(8,4))
sns.boxplot(x = 'area_worst', data=df1, orient='h');
```

#### This function remove the outliers of the features with the medians. Now Data is balanced for the further process.

# 6. Handling Categorical Data

```python
df1=pd.get_dummies(df1,drop_first=True)
df1.head()
```

#### Machine learning models require all input and output variables to be numeric. This means that if your data contains categorical data, you must encode it to numbers before you can fit and evaluate a model. 

# 7. Machine Learning

```python
X=df1.drop(['diagnosis_M'],axis=1)
y=df1['diagnosis_M']
```

> Select X & y
> X=independent columns
> y=Dependent columns(Traget columns)

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

### Stadardization
<blockquote>To comes all Data into one unit excpet Traget columns.Data comes in between(0,1)</blockquote>

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

> Split the Data in Traininig and Testing

```python
print('The shape of X_train is ',X_train.shape)
print('The shape of y_train is ',y_train.shape)
print('The shape of X_test is ',X_test.shape)
print('The shape of y_test is ',y_test.shape)
```

```python

from sklearn.linear_model import LogisticRegression


lr = LogisticRegression()


lr.fit(X_train,y_train)
```

> Model selection
> Import the model
> Instatiate the model
> Fit the model with the data

```python
y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)
```

>Prediction of test

```python
from sklearn import metrics
print('Accuracy of training: ',metrics.accuracy_score(y_train,y_train_pred))
print('Accuracy of testing: ',metrics.accuracy_score(y_test,y_test_pred))
```

<blockquote>Accuraracy of Traing and Testing</blockquote>

#### Confusion Matrix of Training Data.

```python
from sklearn.metrics import confusion_matrix
cfm = confusion_matrix(y_test,y_test_pred)
print(cfm)
```

```python
sns.heatmap(pd.DataFrame(cfm), annot=True,cmap="YlGnBu" ,fmt='g');
plt.tight_layout();
plt.title('Confusion matrix for Training', y=1.1);
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
```

<blockquote>Confusing Metrix Of Traing Data</blockquote>

```python
cnf=confusion_matrix(y_test,y_test_pred)
print(cnf)
```

```python
sns.heatmap(pd.DataFrame(cnf), annot=True,cmap="YlGnBu" ,fmt='g');
plt.tight_layout();
plt.title('Confusion matrix for Testing', y=1.1);
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
```

<blockquote>Confusing Metrix Of Testing Data</blockquote>

```python
from sklearn import metrics
print("Training Precision:",metrics.precision_score(y_train,y_train_pred))
print("Testing Precision:",metrics.precision_score(y_test,y_test_pred))
print("Training Recall:",metrics.recall_score(y_train,y_train_pred))
print("Testing Recall:",metrics.recall_score(y_test,y_test_pred))
```

```python
Training_result=metrics.classification_report(y_train,y_train_pred)
print(Training_result)
```

```python
Testing_result=metrics.classification_report(y_test,y_test_pred)
print(Testing_result)
```

```python
y_pred_proba = lr.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.show()
```

### Logistic regression is an example of supervised learning. It is used to calculate or predict the probability of a binary (yes/no) event occurring. An example of logistic regression could be applying machine learning to determine if a person is likely to be infected with Breast Cancer or not.

### Curve is another common tool used with binary classifiers. The dotted line represents the Receiver operating characteristic (ROC) curve of a purely random classifier; a good classifier stays as far away from that line as possible
